```python
#-*- coding:utf-8 -*-
import urllib3
import json
 
openApiURL = "http://aiopen.etri.re.kr:8000/WiseWWN/WordRel"
accessKey = " 14cb5368-5a38-4c9b-845f-32c4da2befb8 "

names = ['호프', '고깃집', '한식집', '중국집','일식집','탕','분식집','뷔페','카페','통닭집','패스트푸드','경양식집','술집','횟집','라이브카페','김밥','주점','복어','외국음식전문점','냉면집','전통찻집','패밀리레스토랑', '키즈카페','출장조리','다방','이동조리']

tot = []
for firstName in names :
    mat = []
    for secondName in names :    
        firstWord = firstName
        firstSenseId = None
        secondWord = secondName
        secondSenseId = None

        requestJson = {   
            "argument": {
                'first_word': firstWord,
                'first_sense_id': firstSenseId,
                'second_word': secondWord,
                'second_sense_id': secondSenseId
            }
        }

        http = urllib3.PoolManager()
        response = http.request(
            "POST",
            openApiURL,
            headers={"Content-Type": "application/json; charset=UTF-8", "Authorization": accessKey},
            body=json.dumps(requestJson)
        )

       #  print("[responseCode] " + str(response.status))
        # print("[responBody]")
        data = json.loads(response.data)
        
        score = data['return_object']['WWN WordRelInfo']['WordRelInfo']['Similarity'][1]['SimScore']
        mat.append(score)
        # print(str(data))
        # print(score)
    
    print(mat)
    
    tot.append(mat)
print(tot)
        
```
