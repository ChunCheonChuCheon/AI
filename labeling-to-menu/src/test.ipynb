```python
# 기존 데이터프레임에서 특정 컬럼을 선택하여 새로운 데이터프레임을 만들기
import pandas as pd
```

```python
df_main = pd.read_csv('/Users/being/Desktop/[github] workspace/AI/featureSelector/data/c4_general_cafeteria_chuncheon_CC_opened_SS-HJ_addLatiLongti.csv', encoding='utf-8')
```

```python
# 전처리
# '사업장명'은 'cafeteria_name'로 변경
df_main.rename(columns={'사업장명':'cafeteria_name'}, inplace=True)

# '업태구분명'은 'cafeteria_type_orignal'으로 변경
df_main.rename(columns={'업태구분명':'cafeteria_type_orignal'}, inplace=True)

# 'index_cafeteria'는 int형으로 변경
df_main['index_cafeteria'] = df_main['index_cafeteria'].astype(int)

df_main
```

```python
df_manu = pd.read_csv('/Users/being/Desktop/[github] workspace/AI/featureSelector/data/c4_general_cafeteria_chuncheon_CC_opened_SS-HJ_addLatiLongti_menu-price_addedLabel.csv', encoding='utf-8')
```

```python
df_manu
```

```python
# df_manu에서 label열에 결측치가 있는 행을 제거
df_manu_fillter = df_manu.dropna(subset=['label'])
df_manu_fillter
```

```python
# groupby를 통해 'index_cafeteria'을 기준으로 'label'을 묶어준다.(,로 구분)
df_label = df_manu_fillter.groupby('index_cafeteria')['label'].apply(','.join).reset_index()
df_label
```

```python
# df_label의 'label'열의 첫번째 값만 추출하여 리스트로 만들기
label_list = []
for i in df_label['label']:
    label_list.append(i.split(',')[0])
    
df_label['label'] = label_list
```

```python
df_label
```

```python
# cafeteria_name를 기준으로 합치기 how는 합집합
df = pd.merge(df_main, df_label, on='index_cafeteria', how='outer')
df
```

```python
df['label'].value_counts()
```

```python
# label의 널값 분포
df['label'].isnull().sum()
```

```python
# label의 널값을 '기타'로 채우기
df['label'].fillna('기타', inplace=True)
df
```

```python
# 저장하기 to_csv
df.to_csv('/Users/being/Desktop/[github] workspace/AI/featureSelector/data/c4_general_cafeteria_chuncheon_CC_opened_SS-HJ_addLatiLongti_addedLabel_addedLabel.csv', encoding='utf-8', index=True)
```
